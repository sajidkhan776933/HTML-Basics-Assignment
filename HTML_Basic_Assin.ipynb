# Question 1 : Explain the basic structure of an HTML page. Why is it important to follow this structure?

Answer->
*   !DOCTYPE html — Declares the document as an HTML5 document and helps browsers render it correctly.

* html - The root element enclosing all HTML content (except the doctype).

* head - Contains meta-information like page title, character set, and links to stylesheets/scripts; not visible to users.

* title - Shown in the browser tab title bar.

* body - Holds all the content visible to users, including headings, paragraphs, images, links, etc..

# Why Following This Structure Is Important
* Browser Compatibility: Ensures all browsers can display the web page as intended by following universal parsing rules.


* Maintainability: Keeping a standardized structure makes it easier for others (and oneself) to read, update, or manage the web page in the future.

* Error Prevention: Reduces rendering and compatibility errors, helping web pages display content correctly across devices.
